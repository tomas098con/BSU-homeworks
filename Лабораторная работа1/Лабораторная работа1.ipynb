```python
#Задание 1

def rabbits():
    line = input(
        "Введите номер месяца и колличество потомство от одной пары кроликов: "
    ).split()
    n = int(line[0])
    k = int(line[1])
    if n == 0:
        print(f"Колличество пар = 0")
        return
    elif n <= 2:
        print(f"Колличество пар = 1")
        return
    else:
        mounth = 1
        mounth_next = 1
        for _ in range(3, n + 1):
            our_mounth = mounth_next + k * mounth
            mounth = mounth_next
            mounth_next = our_mounth
        print("Колличество пар =", our_mounth)


rabbits()
```

```python
# Задание номер 2


def rabbits2():
    line2 = input("Введите номер месяца и колличество месяцев жизни кроликов: ").split()
    n2 = int(line2[0])
    m = int(line2[1])
    if n2 <= 0 or m <= 0:
        print("Колличество пар = 0")
        return
    elif m == 1:
        print(1 if n2 == 1 else 0)
    else:
        age = [0] * m
        age[0] = 1
        for i in range(1, n2):
            new = sum(age[1:])
            for y in range(m - 1, 0, -1):
                age[y] = age[y - 1]
            age[0] = new
        total = sum(age)
        print(f"Колличество пар: {total}")


rabbits2()
```

```python
# Задание номер 3


def Hamming():
    s = input("Первая цепочка: ").upper()
    t = input("Вторая цепочка: ").upper()

    allowed = set("ACGUT")
    if not set(s).issubset(allowed) or not set(t).issubset(allowed):
        print("Ошибка в буквах!")
        return
    if len(s) != len(t):
        print("Ошибка, цепочки имеют разные длины!")
        return
    else:
        total = 0
        for i in range(0, len(s)):
            if s[i] != t[i]:
                total += 1
        print(f"Расстояние Хэмминга = {total}")


Hamming()
```

```python
# Задание номер 4


def transposition():
    from itertools import permutations

    a = int(input())
    numers = []
    for i in range(1, a + 1):
        numers.append(i)
    transpose = list(permutations(numers))
    print(len(transpose))
    for y in transpose:
        print(y)


transposition()
```

```python
# Задание номер 5


def subsequence():
    from itertools import product

    n = input().strip().split()
    num = int(input().strip())
    for i in product(n, repeat=num):
        print("".join(i))


subsequence()
```

```python
# Задание номер 6
def orientation():
    from itertools import permutations, product

    a = int(input())
    numbers = []
    result = []
    for i in range(1, a + 1):
        numbers.append(i)
    for perm in permutations(numbers):
        for pro in product([-1, 1], repeat=a):
            before_result = list(perm[i] * pro[i] for i in range(a))
            result.append(before_result)
    print(len(result))
    for y in result:
        print(y)


orientation()
```

```python
# Задание номер 7
def Mendel():
    line = input().split()
    k = int(line[0])
    m = int(line[1])
    n = int(line[2])
    num = k + m + n
    ver1 = (n / num) * ((n - 1) / (num - 1))
    ver2 = 0.5 * 2 * (m / num) * (n / (num - 1))
    ver3 = 0.25 * (m / num) * ((m - 1) / (num - 1))
    ver_ = 1 - (ver1 + ver2 + ver3)
    print(ver_)


Mendel()
```
